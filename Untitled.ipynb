```python
renew = data.drop(['self_employed', 'benefits',
       'no_employees', 'remote_work', 'tech_company',
       'care_options', 'wellness_program', 'seek_help',
       'anonymity',  'mental_health_consequence',
       'phys_health_consequence', 'coworkers', 'supervisor',
       'mental_health_interview', 'phys_health_interview',
       'mental_vs_physical', 'obs_consequence'], axis = 1)
renew.head()
```

```python
from sklearn.model_selection import train_test_split

X = renew.drop(['treatment'], axis = 1)
y = renew['treatment']
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=1/4,random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import f1_score
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
pred = model.predict(x_test)
```

```python
acc = accuracy_score(y_test, pred)
confusion_matrix_test = metrics.confusion_matrix(y_test, pred)
recall = metrics.recall_score(y_test, pred)
precision = metrics.precision_score(y_test, pred)
mean_squared_error = metrics.mean_squared_error(y_test, pred)
score = model.score(x_test, y_test)
f1 = f1_score(y_test, pred)
```

```python
print('Accuracy:', acc)
print('confusion_matrix_test:', confusion_matrix_test)
print('recall score:', recall)
print('precision:', precision)
print('mean_squared_error:', mean_squared_error)
print('score:', score)
print('f1 score:', f1)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_forest = RandomForestClassifier()
model_forest.fit(x_train, y_train)
preds = model_forest.predict(x_test)
```

```python
acc1 = accuracy_score(y_test, preds)
confusion_matrix_test1 = metrics.confusion_matrix(y_test, preds)
recall1 = metrics.recall_score(y_test, preds)
precision1 = metrics.precision_score(y_test, preds)
mean_squared_error1 = metrics.mean_squared_error(y_test, preds)
score1 = model.score(x_test, y_test)
f1_1 = f1_score(y_test, preds)
```

```python
print('Accuracy:', acc1)
print('confusion_matrix_test:', confusion_matrix_test1)
print('recall score:', recall1)
print('precision:', precision1)
print('mean_squared_error:', mean_squared_error1)
print('score:', score1)
print('f1 score:', f1_1)
```
